## Dans $\mathbb{R}^3$

1. Écrire une fonction prenant un point $A$ de $\mathbb{R}^3$ et un vecteur $\overrightarrow{u}$ de $\mathbb{R}^3$ et produisant l'équation d'un plan passant par $A$ et orthogonal à $\overrightarrow{u}$.
1. Écrire une fonction prenant un point $A$ de $\mathbb{R}^3$ et un vecteur $\overrightarrow{u}$ de $\mathbb{R}^3$ et produisant des équations de la droite passant par $A$ et dirigée par $\overrightarrow{u}$.

**Question 1**

Si $A=(a,b,)$ et $\overrightarrow{u}=(\alpha,\beta,\gamma)$ l'équation est 
$$\alpha\,(x-a)+\beta\,(y-b)+\gamma\,(z-c)=0$$
Cela donne en `sympy`

```python
import sympy as sp
sp.init_printing()
```

```python
x, y, z = sp.symbols('x y z', real=True)


def plan(A, u):
    global x, y, z
    return sp.Eq(u[0]*(x-A[0])+u[1]*(y-A[1])+u[2]*(z-A[2]), 0)
```

```python
plan([0, 1, 0], [1, 2, 3])
```

**Question 2**

Il y a beaucoup de manières de faire cela. 
1. On peut passer par l'équation paramétrique de la droite
$$x=a+\alpha\,t,\quad y=b+\beta\,t,\quad z=c+\gamma\,t$$
et éliminer le $t$ entre les équations 1 et 2, puis entre 1 et 3 (par exemple). On peut choisir l'équation 1, lorsque $\alpha\ne0$, sinon, il faut changer.
1. On peut poser les deux équations et résoudre le système des équations qui sont obtenus quand on exprime que $A$ est sur la droite et que le vecteur $\overrightarrow{u}$ est dans la direction de la droite.
1. On peut chercher deux vecteurs indépendants, orthogonaux à $\overrightarrow{u}$ et utiliser la question 1.
1. ... et beaucoup d'autres encore !

```python
# Première méthode
def droite(A, u):
    global x, y, z
    if u[0] == 0:
        if u[1] == 0:
            i = 2
        else:
            i = 1
    else:
        i = 0
    j = (i+1) % 3
    k = (i+2) % 3
    M = (x, y, z)
    return (sp.Eq(u[i]*(M[j]-A[j])-u[j]*(M[i]-A[i]), 0), sp.Eq(u[i]*(M[k]-A[k])-u[k]*(M[i]-A[i]), 0))
```

```python
droite([0,1,0], [0,1,2])
```

## Dans $\mathbb{R}^n$

1. Écrire une fonction tirant au hasard un entier $n\in\{2,10\}$, un entier $p\in\{1,n-1\}$, un point $A\in[0,1]^n$ et $p$ vecteurs $\left(\overrightarrow{u}_1,\ldots,\overrightarrow{u}_p\right)$ de $[0,1]^n$ et produisant la ou les équations du sous-espace affine
$$A+\mathrm{Vect}\left(\left\{\overrightarrow{u}_1,\ldots,\overrightarrow{u}_p\right\}\right)$$
1. Améliorer la fonction pour qu'elle vérifie son résultat.

**Question 1**

Quelques remarques avant de commencer.
1. Le nombre d'équations sera 
$$q=n-\underbrace{\mathrm{rang}\left(\left\{\overrightarrow{u}_1,\ldots,\overrightarrow{u}_p\right\}\right)}_{\text{noté }r}$$
1. Le cours de probabilités nous montrera que presque sûrement (probabilité 1) le rang $r$ sera $p$.

Notons $(x_1,\ldots,x_n)$ les coordonnées d'un point quelconque $M\in \mathbb{R}^n$ et $A=(a1,\ldots,a_n)$. L'espace demandé a pour équations paramétriques
$$(*)\qquad\qquad M=A+\sum_{k=1}^p \lambda_k.\overrightarrow{u}_k,\qquad (\lambda_1,\ldots,\lambda_p)\in\mathbb{R}^p$$
Notons $\mathcal{C}^n$ la matrice canonique de $\mathbb{R}^n$, et
$$X=\begin{bmatrix} x_1-a_1\\\vdots\\x_n-a_n\end{bmatrix}=\mathrm{Mat}\left(\overrightarrow{AM},\mathcal{C}^n\right)\in\mathrm{M}_{n,1}(\mathbb{R}),\quad U=[u_{i,j}]_{(i,j)\in\{1,\ldots,n\}\times\{1,\ldots,p\}}=\mathrm{Mat}\left(\left(\overrightarrow{u_1},\ldots,\overrightarrow{u}_p\right),\mathcal{C}^n\right)\in \mathrm{M}_{n,p}(\mathbb{R}),\qquad \Lambda=\begin{bmatrix} \lambda_1\\\vdots\\\lambda_p\end{bmatrix}\in\mathrm{M}_{p,1}(\mathbb{R})$$
La relation $(*)$ sécrit alors $X=U\cdot \Lambda$. On trouve alors, dans le cas où la matrice
$$U_1=[u_{i,j}]_{(i,j)\in\{1,\ldots,p\}}\in \mathrm{GL}_p(\mathbb{R})\text{ que } \Lambda=U_1^{-1}\cdot \begin{bmatrix} x_1-a_1\\\vdots\\ x_p-a_p\end{bmatrix}$$
Les $n-p$ équations cherchées seront alors
$$\begin{bmatrix} x_{p+1}-a_{p+1}\\ \vdots\\ x_n-a_n\end{bmatrix}=[u_{i,j}]_{(i,j)\in\{p+1,\ldots,n\}\times \{1,\ldots,p\}}\cdot U_1^{-1}\cdot \begin{bmatrix} x_1-a_1\\\vdots\\ x_p-a_p\end{bmatrix}$$

```python
import numpy.random as rd
import numpy as np

x = sp.IndexedBase('x')


def solution():
    # Initialisation
    n = rd.randint(2, 11)
    p = rd.randint(1, n)
    U = np.zeros((n, p))
    A = np.zeros(n)
    for i in range(n):
        A[i] = rd.rand()
        for j in range(p):
            U[i, j] = rd.rand()
    U1 = U[:p, :]
    # Calcul
    invU = np.linalg.inv(U1)
    X = np.array([x[i]-A[i] for i in range(n)]).reshape((n, 1))
    M = U[p:, :]@invU@X[:p, :]
    res = [sp.Eq(X[p+k, 0], M[k, 0]) for k in range(n-p)]
    print("L'espace est de dimension "+str(p)+', en voici '+str(n-p)+" équations")
    return res
```

**Question 1**

*On supposera qu'on sait trouver des supplémentaires à tout sous-espace vectoriel de $E$ (ce qui n'est pas évident en général).*

On sait que si $e\in E\setminus H$, on a
$$E=H\oplus\mathbb{K}.e$$
et $H$ est donc de dimension infinie, on peut donc construire une famille libre $(e_i)_{i\in\mathbb{N}}$ de $H$ et supposons connu $H_1$ un sous-espace de $H$ qui vérifie
$$H=\mathrm{Vect}\left(\left\{e_i,\;i\in\mathbb{N}\right\}\right)\oplus H_1$$
On construit alors l'isomorphisme entre $E$ et $H$ de la manière suivante
$$u\;:\left\{\begin{array}{ccc} E&\longrightarrow&H\\ x&\longmapsto&\begin{cases} e_0&\text{si } x=e\\ e_{i+1}&\text{si } x=e_i\\ x&\text{si } x\in H_1\end{cases}\end{array}\right.$$
ou, plus explicitement, si $x\in E$, il existe $\lambda\in\mathbb{K}$, $n\in\mathbb{N}$, $(\mu_1,\ldots,\mu_n)\in\mathbb{K}^n$, $h_1\in H_1$ tel que
$$x=\lambda.e+\sum_{k=1}^n \mu_k.e_{k-1}+h_1$$
et on pose alors
$$u(x)=\lambda.e_0+\sum_{k=1}^n \mu_k.e_k+h_1$$
Le fait que ce soit un isomorphisme est laissé en exercice.

*Le décalage des indices qu'on a utilisé est très fréquent en dimension infinie. Il est bien sûr inutilisable en dimension finie.*
