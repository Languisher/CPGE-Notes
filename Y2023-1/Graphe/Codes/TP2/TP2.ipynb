### 3) Compléments sur les graphes



Quelques fonctions supplémentaires de génération de graphes.

   ```
   subgraph(G, S')          - sous-graphe de G induit par l'ensemble de sommets S'
   union(G1,G2)             - union de graphes
   disjoint_union(G1,G2)    - union de graphes dans le cas où tous les sommets sont différents
   compose(G1,G2)           - combine les graphes en identifiant les noeuds communs au deux
   ```

````python
edgelist1 = [(0, 1), (0, 2), (0, 3)]
edgelist2 = [(0, 4), (0, 5), (0, 6)]
G1 = nx.Graph(edgelist1)
G2 = nx.Graph(edgelist2)
H1 = nx.disjoint_union(G1, G2)
H2 = nx.compose(G1, G2)
print('Sommets de H1 :', H1.nodes)
print('Arêtes de H1 :', H1.edges)
print('Sommets de H2 :', H2.nodes)
print('Arêtes de H2 :', H2.edges)
````

## 1) Quelques rappels sur les dictionnaires

**Syntaxe**. Un dictionnaire s’écrit entre accolades, les couples (clé, valeur) indiqués sous la forme clé:valeur, séparés
par des virgules. Tester les suites de commande suivante :

````python
D={'a': 4, 0:5, (1,2):"coucou"}
````

Contrairement à une liste qui prend en entrée la position, un dictionnaire prend la clé et renvoie la valeur. En particulier si vous cherchez un élément dans une liste cela est linéaire en la taille de la liste. Pour un dictionnaire, la valeur est directement pointée. C'est un gain de temps énorme !!

Les mêmes fonctions que sur les listes existe  : len, pop, append. On peut ajouter la fonction D.keys qui renvoie toutes les clés d'un dictionnaire.

### 3) Gestion des poids

**Méthode n°1**

Plutôt que d'ajouter des arêtes à un graphe, on peut ajouter des arêtes avec poids. Il faut faire attention à ce que les positions des sommets et des labels soient les mêmes. On peut utiliser spring_layout pour le faire automatiquement. Ou bien, on peut spécifier les positions des sommets à l'aide d'un `dictionnaire`.

````python
#Premier graph 

G=nx.Graph()
L=[(1,2,5),(2,3,4),(1,3,5),(1,4,2)]

G.add_weighted_edges_from(L) #On ajoute les arêtes avec leur poid
pos=nx.spring_layout(G) #On doit précise la position pour que les sommets et le label des arêtes doivent être les mettre
edge_labels=nx.get_edge_attributes(G,"weight") #On extrait les différents poids. Essayez sans la commande vous verrez ce qu'on obtient.
nx.draw(G,pos, with_labels=True) #On précise la position pour tracer le graphe
nx.draw_networkx_edge_labels(G,pos,edge_labels   ) #On ajoute le label pour les arêtes.
plt.show() #On trace
````

**Méthode n°2**

Une fois créer le graphe, on crée un dictionnaire avec les différents poids. Tester les commandes suivantes :

````python
G.edges
````

````python
for i,j in G.edges: #On parcourt le graphe
    G[i][j]['weight'] = 1 #On met les poids des arêtes à 1 sous forme de dictionnaire
    
nx.get_edge_attributes(G,'weight') #on attribue les poids à chaque arêtes.
````

Il s'agit bien d'un dictionnaire où chaque arête à un poid de 1.

````python
pos=nx.spring_layout(G) #Comme dans la méthode précédente on précise les positions
weight_labels = nx.get_edge_attributes(G,'weight') #On extrait les poids
nx.draw(G,pos,font_color = 'white',  with_labels = True,) #On trace le graphe
nx.draw_networkx_edge_labels(G,pos,edge_labels=weight_labels) #On précise les labels pour les arêtes.
plt.show() #On trace
````
